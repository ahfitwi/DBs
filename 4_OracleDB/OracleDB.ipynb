### 2.7 Query Filtering Conditions & Operator Precedence

- Parenthesis has the precedence 
***EXAMPLE***: Return those employees that are salesman and that make either 300 dollars in commision or greater than 1000 dollars in commision. 

***Solution***:

    SELECT * FROM emp
    WHERE job = 'SALESMAN'
    AND (COMM = 300 OR COMM > 1000);
     

- ***LIKE*** : equivalent to equal but gives you columns that starts with a cetrain pattern.

        SELECT * FROM emp

        AND (COMM = 300 OR COMM > 1000);
        
        SELECT * FROM emp
        WHERE job LIKE '%MAN'
        AND (COMM = 300 OR COMM > 1000);
