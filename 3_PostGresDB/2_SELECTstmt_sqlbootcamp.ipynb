##### Languages for database querying
- Query language is what allows us to actually take action on databases. It allows us to create, read, update and delete items on our database, as well as more advanced queries like filtering and counting.

- Structured Query Language (SQL) is the most famous of the query languages. SQL grew up alongside the Query By Example (QBE) system developed by IBM in the 1970s. It serves the basis of relational databases.

- With SQL, we can store, retrieve, and manipulate data using simple code snippets, called queries, in an RDBMS (relational database management system).

- The data is stored in the RDBMS in a structured way, where there are relations between the different entities and variables in the data.

- These relations are defined by the database schema, which specifies the relation between various entities and the organization of data for the entities.

##### Purposes of SQL
1. Define Data
    - Create a database with CREATE DATABASE my_database;
    - Create a table with:
        
          CREATE TABLE my_table(
           column1 datatype,
           column2 datatype,
           column3 datatype,
           columnN datatype,
           PRIMARY KEY(columnName)
          );
          
     - Drop a database with 
         - DROP DATABASE my_database;
     - Drop a table with 
         - DROP TABLE my_table;
          
2. Manipulate Data
    - Use a database with USE my_database;
    - Insert data into a database with:

            INSERT INTO my_table (column1, column2, column3,columnN)  
            VALUES (value1, value2, value3,valueN);
    - Select data with: 
    
            SELECT column1, column2, columnN FROM my_table;
    - Select data with a specific condition:
            
            SELECT column1, column2, columnN 
            FROM my_table
            WHERE [condition] // use LIKE, CONTAINS, <, >, etc. here
    - Update query:

            UPDATE my_table
            SET column1 = value1, column2 = value2, columnN = valueN
            WHERE [condition]; 
    - Delete query:

            DELETE FROM my_table
            WHERE [condition];
3. Control Data
    - Transactions:

            COMMIT;
            ROLLBACK;
- SQL is the base for creating structured queries for your relational databases. There are many other “flavors” of SQL that each use SQL in their own way. Different versions of SQL include Oracle PL/SQL, PostgreSQL, and Microsoft Transact-SQL. At a high-level, all are very similar, but each might have their own syntax for certain operations.

### <font color = 'green'>2.3: SELECT DISTINCT</font>

- Sometimes a table contains a column that has duplicate values, and you may find yourself in a situation where you only want to list the unique/distinct values.
- The DISTINCT keyword can be used to return only the distinct values in a column.

        SELECT DISTINCT column(s) FROM table;    
- To clarify which column DISTINCT is being applied to, you can also use parenthesis for clarity:

        SELECT DISTINCT(column(s)) FROM table;
        
- It will work with or with out parenthesis.
- Later on when we learn about adding more calls such as COUNT and DISTINCT together, the parenthesis will be necessary.

        SELECT DISTINCT COUNT(course_id) FROM registrants;

### <font color = 'green'>2.4: COUNT</font>

- The COUNT function returns the number of input rows that match a specific condition of a query.
- Should be the same as the number of rows in the table.
- We can apply COUNT on a specific column or just pass COUNT(\*), we will soon see this should return the same result.
- Syntax:

        SELECT COUNT(course_id) FROM registrants; 
        SELECT COUNT(choice) FROM registrants; 
        SELECT COUNT(*) FROM registrants; 
    - Produce the same result!
- Useful when combined with other keywords:
        
        SELECT COUNT(DISTINCT(choice)) FROM tbl_name; 
        SELECT COUNT(DISTINCT course_id) from registrants;
        SELECT COUNT(DISTINCT choice) FROM tbl_name; 
         

### <font color = 'green'>2.5: SELECT WHERE</font>

- Part of almost evey query you make to the DB
- SELECT and WHERE are the most fundamental SQL staetemtns and you will find yourself using them often.
- The WHERE statement allows use to specify conditions on columns for the rows to be returned.
- Syntax:

        WHERE conditions;
- The WHERE clause appears immediately after the FROM clause of the SELECT statement.
- The condtions are used to filter the rows returned from the SELECT statement.
- PostgreSQL provides a variety of standard operators to construct the conditions. They are available in any SQL Engine.

### <font color = 'green'>2.6: ORDER BY</font>

- You may have noticed PostgreSQL sometimes return the same request query reuslts in a different order.
- You can use ORDER BY to sort rows based on a column value, in either ascending or descending order.
- Syntax:

        SELECT col1, col2
        FROM table
        ORDER BY col1 ASC/DESC
- Notice ORDER BY toward the end of a query, since we want to do any selection and filtering first, before finally sorting.

        SELECT col1, col2
        FROM table
        ORDER BY col1 ASC/DESC
- Use ASC to sort in ascending order 
- Use DESC to sort in descending order
- If you leave it blank ORDER BY uses ASC by default.
- You can also ORDER BY multiple columns. This makes sense when one column has duplicate entries.

        SELECT * FROM customer
        ORDER BY Fname DESC;
        
        SELECT * FROM customer
        ORDER BY Fname DESC, Lname ASC;
    - Tables can be ordered by columns not selected but it is logical to select columns according which ordering is done.

### <font color = 'green'>2.8: BETWEEN</font>

- The BETWEEN operator can be used to match a value against a range of values:
    - value BETWEEN low AND high
        - Inclusivity
- The BETWEEN operator is the same as:
    - value >= low AND value <= high
    - value BETWEEN low AND high.
- You can also combine BETWEEN with the NOT logical operator:
    - value NOT BETWEEN low AND high
        - Exclusivity
- The BETWEEN operator can also be used with dates. Note that you need to format dates in the ISO 8601 standard format, which is YYYY-MM-DD
    - date BETWEEN '2007-01-01' AND '2007-02-01'
    
- Example:

        SELECT * FROM payment
        WHERE amount BETWEEN 9 AND 9;
        
        SELECT * FROM payment
        WHERE amount BETWEEN '2007-02-01' AND '2007-02-15';
        


### <font color = 'green'>2.9: IN</font>

- In a certain cases you want to check for multiple possible value options, for example, if a user's name shows up IN a list of known names.
- We can use the IN operator to create a condition that checks to see if a value in included in a list of multiple options.

        SELECT color 
        FROM table
        WHERE color IN ('red', 'blue')
        
        SELECT color 
        FROM table
        WHERE color NOT IN ('red', 'blue')
        
        SELECT DISTINCT(amount) 
        FROM payment
        WHERE amount IN(0.99, 1.98, 1.99);
        
        SELECT DISTINCT * 
        FROMfrom payment
        WHERE amount IN(0.99, 1.98, 1.99);
        
        SELECT DISTINCT * 
        FROM customer
        WHERE Fname IN('John', 'Jake', 'Julie');
        
        SELECT DISTINCT * 
        FROM customer
        WHERE Fname NOT IN('John', 'Jake', 'Julie');

### <font color = 'green'>2.10: LIKE and ILIKE</font>

- The LIKE operator allows us to perform pattern matching gainst string data with the use of wildcard characters:
    - Percent %
        - Matches any sequence of characters 
    - Underscore_
        - Matches any single chracter
- Examples:
    - All names that begin with an "A"
        - WHERE name LIKE 'A%'
    - All names that end with an 'a'
        - WHERE name LIKE '%a'
- Notice that LIKE is case-sensitive, we can use ILIKE which is case-insensitive.
- Using the underscore allows us to replace just a single character.
    - Get all Mission Impossible films
    - WHERE title LIKE 'Mission IMpossible_'
        - A single wildcard space
- You can use multiple undersocres
- Imagine we had version strng codes in the format 'Version#A4', 'Version#B7', etc
    - WHERE value LIKE 'Version#__'
        - Two spaces
- We can also combine pattern matching operators to create more complex patterns.
    - WHERE name LIKE '_her%'
        - Cheryl
        - Theresa
        - Sherri
- PostGreSQL supports full regex
- Example:

            SELECT COUNT(*)
            FROM customer
            WHERE Fname LIKE 'J%';  --> 65
            
            SELECT *
            FROM customer
            WHERE Fname LIKE '%er%';
            
            SELECT *
            FROM customer
            WHERE Fname NOT LIKE '%er%';
            
            SELECT *
            FROM customer
            WHERE Fname LIKE '%er%';
            
            SELECT * FROM customer
            WHERE Fname NOT LIKE '%er%'
             AND Lname NOT LIKE 'B%'
            ORDER BY Lnam
